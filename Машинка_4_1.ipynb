```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import tree
from sklearn.linear_model import Ridge
from sklearn import linear_model
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import explained_variance_score
from sklearn.metrics import max_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.metrics import classification_report

housing = fetch_california_housing()

x = housing.data[:5000]
y = housing.target[:5000]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
def learning(x_train, y_train, x_test, y_test):
  models = {
      'Linear': LinearRegression(),
      'SVR (rbf)': SVR(kernel='rbf'),
      'SVR (poly)': SVR(kernel='poly'),
      'KNeighbors': KNeighborsRegressor(),
      'MLP': MLPRegressor(max_iter=1000),
      'DecisionTree': tree.DecisionTreeRegressor(),
      'Ridge': Ridge(alpha = 1.0),
      'Lasso': linear_model.Lasso(alpha = 0.1),
      'ElasticNet': ElasticNet(),
      'RandomForest': RandomForestRegressor(),
      'Bagging': BaggingRegressor()
  }
  evs = []
  max_er = []
  mae = []
  mse = []
  r2 = []

  for i, model in models.items():
      model.fit(x_train, y_train)
      y_pred = model.predict(x_test)

      evs.append(explained_variance_score(y_test, y_pred))
      max_er.append(max_error(y_test, y_pred))
      mae.append(mean_absolute_error(y_test, y_pred))
      mse.append(mean_squared_error(y_test, y_pred))
      r2.append(r2_score(y_test, y_pred))


  return pd.DataFrame({"Models" : models.keys(), "Explained_variance_score": evs,
                       "Max_error":max_er, "Mean_absolute_error":mae, "Mean_squared_error":mse, "r2":r2 })
```

```python
learning(x_train, y_train, x_test, y_test)
```
