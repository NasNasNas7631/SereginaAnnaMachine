10. Сколько всего заданий с кратким ответом? С развернутым ответом?

```python
s1 = df1["Задания с кратким ответом"]
s2 = df2["Задания с кратким ответом"]
s3 = df3["Задания с кратким ответом"]
s4 = df4["Задания с кратким ответом"]

def count_plus_signs(text):
    return text.count('+')

def count_minus_signs(text):
    return text.count('-')

print((s1.apply(count_plus_signs)+s1.apply(count_minus_signs)).iloc[0])
print((s2.apply(count_plus_signs)+s2.apply(count_minus_signs)).iloc[0])
print((s3.apply(count_plus_signs)+s3.apply(count_minus_signs)).iloc[0])
print((s4.apply(count_plus_signs)+s4.apply(count_minus_signs)).iloc[0])
```

```python
s1 = df1["Задания с развёрнутым ответом"]
s2 = df2["Задания с развёрнутым ответом"]
s3 = df3["Задания с развёрнутым ответом"]
s4 = df4["Задания с развёрнутым ответом"]

def count_digits(text):
    return sum(c.isdigit() for c in text)//2

print((s1.apply(count_digits)).iloc[0])
print((s2.apply(count_digits)).iloc[0])
print((s3.apply(count_digits)).iloc[0])
print((s4.apply(count_digits)).iloc[0])
```

11. Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
df1 = df1.rename(columns= {'Задания с кратким ответом': 'B'})

a = df1['B'].str.split(pat = '', expand = True)
#g1= a.value_counts(normalize=True) * 100
for i in a:
  k = len(a[a[i]=='+'])
  v = len(a[a[i]=='-'])
  if k!=0 and v!=0:
    print(f'Выполненных заданий в номере {i-1}: {k/3048*100}%')
    print(f'Невыполненных заданий в номере {i-1}: {v/3048*100}%')
```

12. Аналогично и с типом С (ответы с развернутым ответом)

```python
df1 = df1.rename(columns= {'Задания с развёрнутым ответом': 'C'})
df1
```

```python
df1["C"] = df1["C"].str.replace(r"\(.*?\)", "", regex=True)
a = df1['C'].str.split(pat = '', expand = True)

for i in a:
  k = len(a[a[i]!= max(a[i])])
  v = len(a[a[i]==max(a[i])])
  if k!=0 and v!=0:
    print(f'Выполненных заданий в номере {i}: {v/3048*100}%')
    print(f'Невыполненных заданий в номере {i}:{k/3048*100}%')
```
